```python
def save_csv_file(data:pd.DataFrame,file_path:str,file_name:str):
    if not os.path.exists(file_path):
        os.makedirs(file_path)
    csv_file_path = os.path.join(file_path,file_name)
    data.to_csv(csv_file_path,index=False)

```
